```python
import pandas as pd

# Your actual dataframes
table_a = leads.copy()
table_b = internal_db.copy()
table_c = issues.copy()
table_d = appointments.copy()
table_e = contact.copy()

# Use meaningful keys
tables = {
    'leads': table_a,
    'internal_db': table_b,
    'issues': table_c,
    'appointments': table_d,
    'contact': table_e
}

# Function to detect relationships
def find_relationships(tables):
    relationships = []

    for t1_name, t1 in tables.items():
        for t2_name, t2 in tables.items():
            if t1_name == t2_name:
                continue

            for col1 in t1.columns:
                if t1[col1].isnull().all() or t1[col1].dtype not in ['int64', 'object']:
                    continue

                for col2 in t2.columns:
                    if t2[col2].isnull().all() or t2[col2].dtype not in ['int64', 'object']:
                        continue

                    try:
                        match_ratio = t1[col1].isin(t2[col2]).mean()
                    except:
                        continue

                    if match_ratio > 0.95:
                        rel = {
                            'from_table': t1_name,
                            'from_column': col1,
                            'to_table': t2_name,
                            'to_column': col2,
                            'match_ratio': round(match_ratio, 2),
                            'cardinality': get_cardinality(t1, col1, t2, col2)
                        }
                        relationships.append(rel)
    return relationships

# Function to determine cardinality
def get_cardinality(df1, col1, df2, col2):
    u1 = df1[col1].nunique()
    u2 = df2[col2].nunique()

    if u1 == u2 and df1.shape[0] == df2.shape[0]:
        return "1:1"
    elif u1 < u2:
        return "many:1"
    elif u1 > u2:
        return "1:many"
    else:
        return "many:many"

# Run and print results
if __name__ == "__main__":
    results = find_relationships(tables)
    for r in results:
        print(f"{r['from_table']}({r['from_column']}) → {r['to_table']}({r['to_column']}) "
              f"| Match Ratio: {r['match_ratio']} | Cardinality: {r['cardinality']}")
```
